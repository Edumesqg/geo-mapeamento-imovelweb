```python
import pandas as pd
```

```python
data_teste = pd.read_csv('apartment_ads_final.csv')
```

```python
#verificano base de dados
data_teste.head()
```

```python
#excluindo colunas que não serão utilizadas, se existirem
columns_to_drop = ['Title', 'Useful Area', 'Link']
columns_to_drop = [col for col in columns_to_drop if col in data_teste.columns]
data_teste = data_teste.drop(columns=columns_to_drop)
```

```python
#verificando base de dados
data_teste.head()
```

```python
#renomeando colunas
data_teste.rename(columns={'Neighborhood':'bairro', 'Address':'endereco', 'Total Area':'area', 'Bedrooms':'quartos',
                            'Bathrooms':'banheiros', 'Garage':'vaga', 'Rent':'aluguel', 'Condo Fee':'condominio'}, inplace=True)
```

```python
#verificando base de dados
data_teste.head()
```
